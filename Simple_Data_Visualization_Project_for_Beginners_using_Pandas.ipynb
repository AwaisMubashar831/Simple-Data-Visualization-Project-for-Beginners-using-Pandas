#**Step 1: Import Required Libraries**

First, we import the necessary libraries:


*   pandas for data manipulation

* matplotlib.pyplot for visualization



```python
import pandas as pd
import matplotlib.pyplot as plt
```

# **Step 2: Create a Simple Dataset**

Since this is for beginners, we'll manually create a small dataset.

```python
# Create a dictionary with sample sales data
data = {
    "Date": ["2023-01-01", "2023-01-02", "2023-01-03", "2023-01-04", "2023-01-05"],
    "Product": ["Apple", "Banana", "Apple", "Orange", "Banana"],
    "Quantity": [5, 3, 7, 2, 4],
    "Price": [1.5, 0.5, 1.5, 2.0, 0.5]
}

# Convert the dictionary into a Pandas DataFrame
df = pd.DataFrame(data)

# Calculate Total Sales (Quantity * Price)
df["Total_Sales"] = df["Quantity"] * df["Price"]

# Display the DataFrame
print("Sample Sales Data:")
print(df)
```

#**Step 3: Visualizations**

**1.Line Plot - Total Sales Over Time**

We'll plot how sales change over time.

```python
# Convert 'Date' to datetime format for better plotting
df['Date'] = pd.to_datetime(df['Date'])

# Create a line plot
plt.figure(figsize=(8, 4))
plt.plot(df['Date'], df['Total_Sales'], marker='o', linestyle='-', color='blue')

# Add titles and labels
plt.title("Total Sales Over Time")
plt.xlabel("Date")
plt.ylabel("Total Sales ($)")
plt.grid(True)

# Show the plot
plt.show()
```

**Explanation:**


*   Lplt.plot() creates a line graph.

* marker='o' adds dots at each data point.

* plt.grid(True) adds grid lines for better readability.



#**2. Bar Chart - Sales by Product**
We'll compare total sales for each product.

```python
# Group data by Product and sum the Total_Sales
product_sales = df.groupby('Product')['Total_Sales'].sum()

# Create a bar chart
plt.figure(figsize=(8, 4))
product_sales.plot(kind='bar', color=['red', 'yellow', 'orange'])

# Add titles and labels
plt.title("Total Sales by Product")
plt.xlabel("Product")
plt.ylabel("Total Sales ($)")
plt.xticks(rotation=0)  # Keep product names horizontal

# Show the plot
plt.show()
```

**Explanation:**


*   groupby() groups data by 'Product' and sums 'Total_Sales'.

*  kind='bar' creates a bar chart.



#**3. Histogram - Price Distribution**
We'll see how prices are distributed.

```python
plt.figure(figsize=(8, 4))
plt.hist(df['Price'], bins=5, color='green', edgecolor='black')

# Add titles and labels
plt.title("Distribution of Prices")
plt.xlabel("Price ($)")
plt.ylabel("Frequency")

# Show the plot
plt.show()
```

#**Explanation:**


*   plt.hist() creates a histogram.

* bins=5 divides data into 5 ranges.m



#**4. Scatter Plot - Quantity vs. Total Sales**
We'll check if higher quantities lead to higher sales.

```python
plt.figure(figsize=(8, 4))
plt.scatter(df['Quantity'], df['Total_Sales'], color='purple')

# Add titles and labels
plt.title("Quantity vs. Total Sales")
plt.xlabel("Quantity Sold")
plt.ylabel("Total Sales ($)")

# Show the plot
plt.show()
```

**Explanation:**


*   plt.scatter() creates a scatter plot.

*  Helps identify trends (e.g., more quantity â†’ higher sales).
